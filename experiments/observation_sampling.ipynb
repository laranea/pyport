Goal of this notebook is to solve the problem of turning daily data into lower frequencies. That is, take daily data for say, the opening prices of facebook and find the weekly opening price. Do the same for each of the prices. Maybe also throw in volumn there for good measure...


```python
# let's get some data to play with.
import time
import pandas as pd
import pandas_datareader as pdr
from findatapy.market import Market, MarketDataGenerator, MarketDataRequest
from findatapy.util import SwimPool; SwimPool()

market = Market(market_data_generator=MarketDataGenerator())

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
data_request = MarketDataRequest(
    data_source='yahoo',
    start_date='year',
    tickers=['FB', 'GOOGL'],
    fields=['open', 'close', 'high','low'],
    vendor_tickers=['FB','GOOGL'],
    vendor_fields=['Open','Close', 'High','Low'])
df = market.fetch_market(data_request)
df.tail()
```

Compare tha with the normal pandas data reader.

```python
data_request = MarketDataRequest(
    data_source='yahoo',
    start_date='decade',
    tickers=ticker_list,
    fields=['open', 'close', 'high','low', 'adjclose'],
    vendor_tickers=ticker_list,
    vendor_fields=['Open','Close', 'High','Low','Adj Close'])
df = market.fetch_market(data_request)
df.tail()
```
